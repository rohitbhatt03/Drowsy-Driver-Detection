```python
import torch 
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
model=torch.hub.load('ultralytics/yolov5','yolov5s')
```

```python
img='https://media.istockphoto.com/id/159406920/photo/aerial-view-of-cars-in-traffic.jpg?b=1&s=170667a&w=0&k=20&c=OWuBJE9s21eKyBtCznr6HWeALNJEBZs1Lk8O83gO5w0='
```

```python
results=model(img)
results.print()
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()
```

```python
cap=cv2.VideoCapture(0)
while cap.isOpened():
    ret,frame=cap.read()
    
    result=model(frame)
    
    cv2.imshow('balak',np.squeeze(result.render()))
    
    if cv2.waitKey(10) & 0xFF==ord('q'):
        break 
cap.release()
cv2.destroyAllWindows()
```

```python
import uuid
import os
import time
```

```python
IMAGES_PATH=os.path.join('data','images')
labels=['awake','drowsy']
number_img=20
```

```python
cap=cv2.VideoCapture(0)
for label in labels:
    print('collectiong images for {}'.format(label))
    time.sleep(5)
    for img in range(number_img):
        print('collecting images for {},image number {} '.format(label,img))
        #web cam feed
        ret,frame=cap.read()
        #Naming out image path
        
        imgname = os.path.join(IMAGES_PATH,label+'.'+str(uuid.uuid1())+'.jpg')
        #write out image to file
        cv2.imwrite(imgname, frame)
        
        
        cv2.imshow('Image Collection', frame)
        time.sleep(2)
        if cv2.waitKey(10) & 0xFF==ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
